```python
from sklearn import manifold
from matplotlib.collections import LineCollection


def double_centering(M):
    
    N = M.shape[0]

    Mc = - 0.5*( M - 1./N * np.sum(M, axis = 0)[None, :] - 1./N * np.sum(M, axis =1)[:, None] + 1./N**2 * np.sum(M) )
    
    return Mc
```

```python
from sklearn.metrics.cluster import normalized_mutual_info_score, adjusted_mutual_info_score, mutual_info_score
from sklearn.metrics.cluster import adjusted_rand_score, homogeneity_score, fowlkes_mallows_score
```

```python
NMIs_gt = []
NMIs_a = []

for d in ds.data_sets:
    NMI = normalized_mutual_info_score(d.labels[:n_an_art], gtl[:n_an_art], average_method='arithmetic')
    NMI = adjusted_rand_score(d.labels[:n_an_art], gtl[:n_an_art])
    NMIs_gt.append(NMI)

    NMI = normalized_mutual_info_score(d.labels[:n_an_art], only_ancestors_num_012[:n_an_art], average_method='arithmetic')
    NMI = adjusted_rand_score(d.labels[:n_an_art], only_ancestors_num_012[:n_an_art])

    adjusted_rand_score
    NMIs_a.append(NMI)
```

```python

fig, axs = plt.subplots(1, 4, figsize=(8, 2), sharey=False, gridspec_kw={'width_ratios': [1, 1, 1, 1.25]})

for li, layer in enumerate([0, 34, 142, 153]):
    
    cp = np.where(np.array(layers) == layer )[0][0]
    
    d = ds.data_sets[cp]
    
    an_fracs = []

    for i in range(d.Nclus_m):
        idx = d.clstruct_m[i]

        ancs = only_ancestors_num_012[idx]
        n_animals = sum(ancs == 13)
        n_artifacts = sum(ancs == 2)

        an_frac = n_animals/(n_animals + n_artifacts) - (38400/(47100 + 38400)) + 0.5
        an_fracs.append(an_frac)
    
    # generate distance matrix as max density minus the border density
    d_dis = max(d.Rho) - d.out_bord 
    d_dis[np.diag_indices_from(d_dis)] = 0.

    md = np.mean(d_dis[np.triu_indices_from(d_dis)])
    sig = np.mean(md)
    ker = np.exp(- d_dis**2/(0.5*sig**2))#Rho_bord_m
    ker = double_centering(-ker)
    #ker = double_centering(d_dis**2)
    e, v = np.linalg.eigh(ker)
    s = np.argsort(e)[::-1]
    out = v[:, s[:2]]


    s = []
    col = []
    for i in range(d.Nclus_m):
        s.append(3. * np.sqrt(len(d.clstruct_m[i])))
        col.append(cm.coolwarm(an_fracs[i]))

    s = np.array(s)
    col = np.array(col)
    
    if layer == 34:
        axs[li].scatter(0, 0, s=s, c=col, edgecolor = 'k')
        
    elif layer == 153:
        top = 235
        sc = axs[li].scatter(out[:top, 0], out[:top, 1], s=s[:top], c=col[:top], edgecolor = 'k')
        # cax = make_square_axes_with_colorbar(axs[li], size=0.15, pad=0.1)
        #fig.colorbar(sc, cax=cax)
    else:
        
        axs[li].scatter(out[:, 0], out[:, 1], s=s, c=col, edgecolor = 'k')
        

    # # write cluster number inside cluster
    # for i in range(d.Nclus_m):
    #     ax.annotate(i, (out[i, 0], out[i, 1]))

    if layer == 153:
        cbar = plt.colorbar(cm.ScalarMappable(cmap = cm.coolwarm))
        cbar.set_label(r'$ \text{artifacts} \rightarrow \text{animals}$', rotation=90)

titles = [r'$l=0$', r'$l=34$', r'$l=142$', r'$l=153$']

for ax, title in zip(axs, titles):
    ax.set_title(title)
    ax.set_xticks([],[])
    ax.set_yticks([],[])
    
    ax.set_xlabel(r'$c_1$')
    ax.set_ylabel(r'$c_2$')


plt.savefig('c{}.png'.format(cp), dpi = 300)
```

```python
fig = plt.figure(figsize=(9.5, 3), constrained_layout=True)

widths = [1, 1, 1, 1, 1, 1]
heights = [1., 1.3]
gs = fig.add_gridspec(nrows=2, ncols=6, width_ratios=widths, height_ratios=heights)

titles = [r'$l=0$', r'$l=34$',  r'$l=97$' ,r'$l=142$',r'$l=145$', r'$l=153$']


for li, layer in enumerate([0, 34, 97, 142, 145, 153]):
    
    ax_li = fig.add_subplot(gs[0, li])

    cp = np.where(np.array(layers) == layer )[0][0]
    d = ds.data_sets[cp]
    
    an_fracs = []

    for i in range(d.Nclus_m):
        idx = d.clstruct_m[i]

        ancs = only_ancestors_num_012[idx]
        n_animals = sum(ancs == 13)
        n_artifacts = sum(ancs == 2)

        an_frac = n_animals/(n_animals + n_artifacts) - (38400/(47100 + 38400)) + 0.5
        an_fracs.append(an_frac)
    
    # generate distance matrix as max density minus the border density
    d_dis = max(d.Rho) - d.out_bord 
    d_dis[np.diag_indices_from(d_dis)] = 0.

    md = np.mean(d_dis[np.triu_indices_from(d_dis)])
    sig = 1.*np.mean(md)
    ker = np.exp(- d_dis**2/(0.5*sig**2))#Rho_bord_m
    ker = double_centering(-ker)
    #ker = double_centering(d_dis**2)
    e, v = np.linalg.eigh(ker)
    s = np.argsort(e)[::-1]
    out = v[:, s[:2]]


    s = []
    col = []
    for i in range(d.Nclus_m):
        s.append(3. * np.sqrt(len(d.clstruct_m[i])))
        col.append(cm.coolwarm(an_fracs[i]))

    s = np.array(s)
    col = np.array(col)
    
    if layer == 34:
        ax_li.scatter(0, 0, s=s, c=col, edgecolor = 'k')
        
    elif layer == 97:
        theta = np.radians(-120)
        
        c_, s_ = np.cos(theta), np.sin(theta)
        R = np.array(((c_, -s_), (s_, c_))) 
        out_ = (np.dot(R, out.T)).T
        
        ax_li.scatter(out_[:, 0], out_[:, 1], s=s, c=col, edgecolor = 'k')
    
    
    elif layer == 142:
        theta = np.radians(30)
        
        c_, s_ = np.cos(theta), np.sin(theta)
        R = np.array(((c_, -s_), (s_, c_))) 
        out_ = np.dot(np.diag([1, -1]) ,np.dot(R, out.T) ).T
        
        ax_li.scatter(out_[:, 0], out_[:, 1], s=s, c=col, edgecolor = 'k')
      
    
    elif layer == 145:
        theta = np.radians(170)
        
        c_, s_ = np.cos(theta), np.sin(theta)
        R = np.array(((c_, -s_), (s_, c_))) 
        out_ = (np.dot(R, out.T)).T
        
        ax_li.scatter(out_[:, 0], out_[:, 1], s=s, c=col, edgecolor = 'k')
        
        
    elif layer == 153:
        top = 215
        
        # rotate data
        theta = np.radians(-160)
        
        c_, s_ = np.cos(theta), np.sin(theta)
        R = np.array(((c_, -s_), (s_, c_))) 
        out_ = (np.dot(R, out[:top, :2].T)).T
        
        print(out_.shape)
        ax_li.scatter(out_[:, 0], out_[:, 1], s=s[:top], c=col[:top], edgecolor = 'k')
        
    else:        
        ax_li.scatter(out[:, 0], out[:, 1], s=s, c=col, edgecolor = 'k')

    
    ax_li.set_xlabel(titles[li])
    ax_li.set_xticks([],[])
    ax_li.set_yticks([],[])
    
ax5 = fig.add_subplot(gs[1, :])   

#ax5.plot(layers, np.array(label_overlaps)*.6, '--', label = r'$\chi^{gt}$', c = 'grey', marker = 'o', markersize = 6,rasterized=True)

ax5.plot(layers, NMIs_gt, '-o', label = 'ARI classes (goldfish, car wheel, ...)', marker='o', rasterized=True, markersize = 6)
ax5.plot(layers, NMIs_a, '-o', label = 'ARI macro-classes (animals, artifacts, ...)', marker = 'o', rasterized=True, markersize = 6)


ax5.set_xlim(-2, 156)
ax5.set_yticks([0, 0.2, 0.4, 0.6])
ax5.set_xlabel(r'$\text{layer } (l)$')
ax5.set_ylabel(r'$\text{ARI}$')
ax5.legend()
ax5.set_ylim([-.05, .6])

#ax52 = ax5.twinx() 
#ax52.grid(None)
#ax52.set_ylabel(r'$\chi^{gt}$')
#ax52.set_ylim([-.05/.6, .6/.6])

plt.savefig('clustering.png', dpi = 600)
plt.show()
```
